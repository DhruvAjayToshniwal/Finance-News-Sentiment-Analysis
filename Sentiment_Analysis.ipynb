```python
import requests
import json
import openai
from transformers import pipeline
import pandas as pd
import plotly.express as px

# Define function to collect and analyze articles
def analyze_articles(query):
    # Get articles
    url = 'https://gnews.io/api/v4/search?'
    parameters = {
        'q': query,
        'token': '',#add your token here
        'lang': 'en',
        'country': 'us'
    }
    response = requests.get(url, params=parameters)
    data = response.json()
    articles = data['articles']

    # Set OpenAI key
    openai.api_key = '' #use openai api key

    # Initialize sentiment analyzer
    sentiment_analyzer = pipeline('sentiment-analysis')

    df = pd.DataFrame(columns=['title', 'summary', 'sentiment', 'score'])

    # Summarize and analyze sentiment of each article
    for i, article in enumerate(articles):
        prompt = "Summarize: " + article['content']
        response = openai.Completion.create(
            engine="text-davinci-002",
            prompt=prompt,
            temperature=0.3,
            max_tokens=100
        )
        summary = response['choices'][0]['text'].strip()
        result = sentiment_analyzer(summary)
        sentiment = 1 if result[0]['label'] == 'POSITIVE' else -1  # Assign +1 for positive sentiment and -1 for negative
        score = result[0]['score']
        df.loc[i] = [article['title'], summary, sentiment, score]

        print(f"Title: {article['title']}\nSummary: {summary}\nSentiment: {sentiment}, Score: {score}\n")

    # Plot sentiment over time (order of articles)
    fig = px.line(df, y='sentiment', labels={'index': 'Articles (from most recent)', 'sentiment': 'Sentiment Score'}, title='Sentiment Analysis of Finance News Over Time')
    fig.show()

# Define queries
queries = ['finance', 'economy', 'bitcoin']

# Run analysis for each query
for query in queries:
    analyze_articles(query)
```
